```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load anh
img = plt.imread('pic.jpg', cv2.IMREAD_UNCHANGED)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print('image shape:', img.shape)
print('gray shape:', gray.shape)
```

```python
#hien thi anh
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(gray)
plt.title('Gray Image')
```

```python
#tinh toan cac dac trung
#1. Gradient Gx, Gy
gx = cv2.Sobel(gray, cv2.CV_32F, dx = 0, dy = 1,ksize = 3)
gy = cv2.Sobel(gray, cv2.CV_32F, dx = 1, dy = 0, ksize = 3)
print('gray shape: {}'.format(gray.shape))
print('gx shape: {}'.format(gx.shape))
print('gy shape: {}'.format(gy.shape))
#2. Tinh G va phuong Gradient Theta.
g, theta = cv2.cartToPolar(gx, gy, angleInDegrees = True)
print('gradient format: {}'.format(g.shape))
print('theta format: {}'.format(theta.shape))
```

```python
#3. Hien thi gx, gy, g, theta
w = 20
h = 10
plt.figure(figsize=(w,h))

plt.subplot(1,5,1)
plt.title('gx')
plt.imshow(gx)

plt.subplot(1,5,2)
plt.title('gy')
plt.imshow(gy)

plt.subplot(1,5,3)
plt.title('Gradient Magnitude')
plt.imshow(g)

plt.subplot(1,5,4)
plt.title('Gradient Direction')
plt.imshow(theta)

plt.subplot(1,5,5)
plt.title('Origin')
plt.imshow(img)
```
